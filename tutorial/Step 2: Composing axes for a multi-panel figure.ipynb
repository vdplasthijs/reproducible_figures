## Composing ax structure using gridspec:
In the example above, we have used `matplotlib`'s most basic function for composing axes: `plt.subplots(nrows=1, ncols=1, ..)`, which creates a grid of equal-sized panels (with number of rows and columns as specified by the function args). Though a very useful function, it is essentially a shortcut (for creating equal-sized panels) of `matplotlib`'s fundamental axes composition function: `gridspec`. Therefore, we will focus on the underlying basics of creating _any_ composition of axes using `gridspec`. Once this is clear, it can be useful going back to shortcuts like `plt.subplots()` if that suits your use-case. 

First, it's important to understand the difference between a _figure_ (of type `mpl.figure.Figure`) and an _axis_ (of type `mpl.axes._subplots.AxesSubplot`). Axes occupy the figure.

```python
## Print types of previously defined figure 'fig' and axis 'ax'
print(type(fig), type(ax_1))
```

By default, `plt.subplots()` creates a grid of axes that together occupy the entire figure. In contrast, using `gridspec`, we can define what part of the figure should be occupied by axes by using the `bottom`, `top`, `left` and `right` limits. Further, we can add multiple gridspec to the same figure, enabling complex compositions:

```python
## Make figure:
fig = plt.figure(figsize=(10, 8))  # width x height in inches
## Add a grid of subplots:
gs_1 = fig.add_gridspec(ncols=2, nrows=2,  # number of columns and rows
                      bottom=0.05, left=0.05, top=0.7, right=0.5)  # the limits of the grid
gs_2 = fig.add_gridspec(ncols=1, nrows=1,  # number of columns and rows
                        bottom=0.85, left=0.05, top=0.98, right=0.95)  # the limits of the grid

## Gridspec can be then used to create axes:
ax_1 = fig.add_subplot(gs_1[0, 0])
ax_2 = fig.add_subplot(gs_1[0, 1])
ax_3 = fig.add_subplot(gs_1[1, 0])
ax_4 = fig.add_subplot(gs_1[1, 1])

ax_5 = fig.add_subplot(gs_2[0, 0])

plt.show()
```

Other notable arguments of `gridspec` are: `hspace` for controlling the height-space (i.e., vertical space) between panels within that grid, and `wspace` for controlling the width-space. 
Additionally, `width_ratios` and `height_ratios` control the ratios of width/height across panels in the grid.
The example below uses different `wspace` and `hspace` for `gs_1`, and splits `gs_2` up into 4 panels with varying `width_ratio`:

```python
## Make figure:
fig = plt.figure(figsize=(10, 8))  # width x height in inches
## Add a grid of subplots:
gs_1 = fig.add_gridspec(ncols=2, nrows=2,  # number of columns and rows
                       bottom=0.05, left=0.05, top=0.7, right=0.5,  # the limits of the grid
                       wspace=0.4, hspace=0.05)  # the amount of width/height reserved for space between subplots

gs_2 = fig.add_gridspec(ncols=4, nrows=1,  # number of columns and rows
                        bottom=0.85, left=0.05, top=0.98, right=0.95,  # the limits of the grid
                        width_ratios=[1, 2, 3, 5])

## Gridspec can be then used to create axes:
ax_1 = fig.add_subplot(gs_1[0, 0])
ax_2 = fig.add_subplot(gs_1[0, 1])
ax_3 = fig.add_subplot(gs_1[1, 0])
ax_4 = fig.add_subplot(gs_1[1, 1])

ax_5 = fig.add_subplot(gs_2[0, 0])
ax_6 = fig.add_subplot(gs_2[0, 1])
ax_7 = fig.add_subplot(gs_2[0, 2])
ax_8 = fig.add_subplot(gs_2[0, 3])

plt.show()
```
